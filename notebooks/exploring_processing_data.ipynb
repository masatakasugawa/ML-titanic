```python
# imports
import pandas as pd
import numpy as np
import os
```

```python
#set the path of the raw data
raw_data_path = os.path.join(os.path.pardir, 'data', 'raw')
train_file_path = os.path.join(raw_data_path, 'train.csv')
test_file_path = os.path.join(raw_data_path, 'test.csv')
```

```python
#read the data with all defalut parameters
train_df = pd.read_csv(train_file_path, index_col='PassengerId')
test_df = pd.read_csv(test_file_path, index_col='PassengerId')
```

```python
#get the type
type(train_df)
```

```python
#use .info() to get brief infomation about the dataframe
train_df.info()
# Survived -> if survived 1-yes/o-no
# Pclass   -> passenger classs 1-1st class/2-2nd class/3-3rd class
# Name     -> passenger name
# Sex      -> Gender (Male or female?)
# Age      -> passenger age
# SibSp    -> NUmber of siblings / spouses aboard(兄弟や配偶者の数)
# Parch.   -> Number of Parents / children aboard(親または子供の数)
# Ticket   -> ticket number
# Fare     -> passenger fare(料金)
# Cabin    -> cabin(?)
# Embarked -> Point of embarkment(C=Cherbourg Q=Queenstown S=Southampton)(乗船した地名)
```

```python
test_df.info()
```

```python
test_df['Survived'] = -888 # adding survived with a defalut value
```

```python
df = pd.concat((train_df, test_df))
```

```python
# Filter rows based on the condition
male_passengers = df.loc[df.Sex == 'male', :]
print(type(male_passengers))
print('Number of male passengers : {0}'.format(len(male_passengers)))
```

```python
# use & or | operators to build complex logic
male_passengers_firest_class = df.loc[((df.Sex == 'male') & (df.Pclass == 1)), :]
print(type(male_passengers_firest_class))
print('Number of male passengers in first class : {0}'.format(len(male_passengers_firest_class)))
```

```python
# use .discribe() to get statistics for all numeric columns
df.describe()
```

```python
# numerical feature
# centrality measures
print('Mean fare : {0}'.format(df.Fare.mean())) # mean：平均値
print('Median fare : {0}'.format(df.Fare.median())) #median：中央値
```

```python
# disperson measures
print('Min fare : {0}'.format(df.Fare.min())) # min：最小値
print('Max fare : {0}'.format(df.Fare.max())) # max：最大値
print('Fare range : {0}'.format(df.Fare.max() - df.Fare.min())) # range：範囲
print('Fare 25 percentile : {0}'.format(df.Fare.quantile(.25))) # 25 Percentile：四方分位１〜２５％
print('Fare 50 percentile : {0}'.format(df.Fare.quantile(.5))) # 50 percentile：四方分位２６〜５０％
print('Fare 75 percentile : {0}'.format(df.Fare.quantile(.75))) # 75 percentile：四方分位５１〜７５％
print('Variance fare : {0}'.format(df.Fare.var())) # variance：差異
print('Standard deviation fare : {0}'.format(df.Fare.std())) # standard deviation：標準偏差
```
